```python
#
bars = ('Especializada','Central')
x_pos = np.arange(len(bars))
er_bib.TipoBib.value_counts().plot(kind='bar')
plt.xticks(x_pos, bars)
plt.xlabel('Tipo de biblioteca', labelpad=14)
plt.ylabel('Cantidad', labelpad=14)
plt.title('Bibliotecas en universidades públicas según tipo', y=1.02)
plt.show;
```
